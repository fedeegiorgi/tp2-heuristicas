```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
data = {}

with open('output_real.csv', 'r') as file:
    reader = csv.DictReader(file)
    row = next(reader)
    for column, value in row.items():
        if column != 'instance' and column != 'type_instance': 
            value = float(value) 
            data[column] = value
        if column == 'n' or column == 'm':
            value = int(value) 
            data[column] = value

print(data)
```

```python
# Define the data
categories = ['ov_mincost', 'ov_mincost_relocate', 'ov_mincost_swap','ov_mincost_relocate_swap','ov_mincost_swap_relocate']
values = [data['ov_mincost'], data['ov_mincost_relocate'], data['ov_mincost_swap'], data['ov_mincost_relocate_swap'], data['ov_mincost_swap_relocate']]

# Define colors for the bars
colors = ['#d24234', '#e09142', '#e8c943', '#376da5', '#3c815e']

# Create an array of indices for the x-axis positions
x = np.arange(len(categories))

# Set the width of the bars
width = 0.5
label_font = {'fontname': 'Times New Roman', 'fontsize': 12, 'fontweight': 'medium'}
# Create the bar plot with custom width and colors
plt.bar(x, values, width=width, color=colors, label=categories)

for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Set the plot title and axis labels
plt.title('Comparación solución en base a MinCostHeuristic',**label_font)
plt.xlabel('Heuristica',**label_font)
plt.ylabel('Valor Objetivo',**label_font)

# Set the x-axis tick labels
plt.xticks([], [])

# Display the plot
plt.legend(loc='upper right', bbox_to_anchor=(1.45, 1.0), prop= 'Times New Roman')
plt.show()
```
